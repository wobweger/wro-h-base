```python
oSoup=BeautifulSoup(html.content,'html.parser')
```

```python
oSoup.find('table').findAll('td',class_='TTRow')
```

```python
rows=[]
lRow=oSoup.find('table').findAll('tr')
for oRow in lRow:
    lCell=oRow.findAll('td')
    B=[]
    for i in range(0,min(5,len(lCell))):
        B.append(lCell[i].text.replace('\n',''))
    rows.append(B)
oB=pd.DataFrame(rows)
```
